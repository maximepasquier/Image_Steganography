```python
import numpy as np
import skimage
import os
import cv2
import scipy.signal
import random
import math
import matplotlib.pyplot as plt
```

```python
def Mean_Square_Error(X,Y):
    return (np.square(X - Y)).mean()
```

```python
def Peak_Signal_to_Noise_Ratio(X,Y):
    mse = Mean_Square_Error(X,Y)
    if(mse == 0):
        return 100
    max_pixel = 256
    psnr = 20 * math.log10(max_pixel / math.sqrt(mse))
    return psnr
```
